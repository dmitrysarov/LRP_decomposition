```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
inpt = np.random.randint(0, 255, size=(10, 10))
w = np.random.randn(10, 10)
prod = np.sum(w*inpt) #considered transformation
print(prod)
```

```python
fig, ax = plt.subplots(1,2)
im1=ax[0].imshow(inpt)
fig.colorbar(im1,ax = ax[0])
im2=ax[1].imshow(w)
fig.colorbar(im2,ax = ax[1])

plt.show()
```

# w2 rule

#### root

```python
root = inpt - w*np.sum(w*inpt)/(w**2).sum() #max gradient direction 
```

```python
prod_2 = np.sum(w*root)
print(prod_2)
```

```python
plt.hist(inpt.flatten())
plt.show()
```

```python
plt.hist(root.flatten())
plt.show()
```

```python
fig=plt.imshow(root)
```

#### R update

```python
Ri = w**2/sum(w**2)*prod
```

```python
plt.hist(Ri.flatten())
plt.show()
```

```python
plt.imshow(Ri)
plt.colorbar()
```

# Z+ rule

#### root

```python
seqrch_dir = np.zeros_like(w)
seqrch_dir[w>0] = inpt[w>0]
root_2 = inpt - seqrch_dir/(seqrch_dir*w).sum()*np.sum(w*inpt) #restricted search domain (positive weight*inpt)

prod_4 = np.sum(w*root_2)
print(prod_4)
```

```python
wp = np.zeros_like(w)
wp[w>0] = w[w>0]
Ri_2 = wp*inpt/sum(wp*inpt)*prod
plt.imshow(Ri_2)
plt.colorbar()
```
